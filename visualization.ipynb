```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [21, 345,56,12,23,453,64,1]
plt.plot(y)
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.pie(y)
plt.show()
```

```python
fig , ax = plt.subplots(nrows=2, ncols=1, figsize= (15,5))
plt.show()
```

```python
t = np.arange(-3 * np.pi, 3* np.pi , 0.2)
x = np.sin(t)
y = np.cos(t)
fig, ax = plt.subplots(nrows=2, ncols=2 , figsize=(15,2))
ax[0,][0].plot(t , x , 'r')
ax[0][1].bar(t , y)
plt.show
```
